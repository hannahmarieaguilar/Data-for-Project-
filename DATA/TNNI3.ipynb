```python
from datascience import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

import pandas as pd
import warnings
warnings.simplefilter("ignore")
```

```python
TNNI3_Path = pd.read_csv('TNNI3.txt', sep='\t').dropna(subset=['Protein change'])
TNNI3_Path
```

```python
TNNI3_LP = pd.read_csv('TNNI3_LP.txt', sep='\t')
newTNNI3_LP = TNNI3_LP['Condition(s)'].str.contains("Hypertrophic cardiomyopathy")
TNNI3_LP1 = TNNI3_LP[newTNNI3_LP].dropna(subset=['Protein change'])
TNNI3_LP1
```

```python
combined2 = [TNNI3_LP1, TNNI3_Path]
TNNI3_combined = pd.concat(combined2)
TNNI3_combined
```

```python
combo = TNNI3_combined[['Name']]
np.array(np.array(combo['Name'].str.split(' ')).item(0)).item(0)
lst = []
for i in np.arange(len(TNNI3_combined)):
    lst = np.append(lst, np.array(np.array(combo['Name'].str.split(' ')).item(i)).item(0))
    
lst
combo = Table().with_columns('Name', lst)
combo.show()
```

```python
import numpy as np
TNNI3_ensembl_np = Table().read_table('TNNI3_ensembl.txt', sep='\t').where('BIOTYPE', 'protein_coding')
new = TNNI3_ensembl_np.select('#Uploaded_variation', 'Consequence')
new
```

```python
def label_consequence(consequence):
    if 'stop_gained' in consequence:
        return 'stop_gained'
    elif 'missense_variant' in consequence:
        return 'missense_variant'
    elif 'frameshift_variant' in consequence:
        return 'frameshift_variant'
    elif 'inframe_deletion' in consequence:
        return 'inframe_deletion'
```

```python
try1 = new.apply(label_consequence, 'Consequence')
new2 = new.with_column('Our Consequences', try1)
new2
yay = new2.where('Our Consequences', are.not_equal_to(None))
test3 = yay.group('Our Consequences')
test3
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.pie(test3.column('count'), autopct='%1.1f%%', shadow = True)
#,labels = test3.column('Our Consequences'), autopct='%1.1f%%', shadow = True)
plt.legend(title = 'TNNI3 Consequences:', fontsize="9", labels = test3.column('Our Consequences'), loc ="lower center", bbox_to_anchor=(0.37, -0.2, 0.2, 0.2))
```
