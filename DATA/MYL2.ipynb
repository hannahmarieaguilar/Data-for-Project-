```python
from datascience import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

import pandas as pd
import warnings
warnings.simplefilter("ignore")
```

```python
MYL2_Path = pd.read_csv('MYL2.txt', sep='\t').dropna(subset=['Protein change'])
MYL2_Path
```

```python
MYL2_LP = pd.read_csv('MYL2_LP.txt', sep='\t')
newMYL2_LP = MYL2_LP['Condition(s)'].str.contains("Hypertrophic cardiomyopathy")
MYL2_LP1 = MYL2_LP[newMYL2_LP].dropna(subset=['Protein change'])
MYL2_LP1
```

```python
combined2 = [MYL2_LP1, MYL2_Path]
MYL2_combined = pd.concat(combined2)
MYL2_combined
```

```python
combo = MYL2_combined[['Name']]
np.array(np.array(combo['Name'].str.split(' ')).item(0)).item(0)
lst = []
for i in np.arange(len(MYL2_combined)):
    lst = np.append(lst, np.array(np.array(combo['Name'].str.split(' ')).item(i)).item(0))
    
lst
combo = Table().with_columns('Name', lst)
combo.show()
```

```python
import numpy as np
MYL2_ensembl_np = Table().read_table('MYL2_ensembl.txt', sep='\t').where('BIOTYPE', 'protein_coding')
new = MYL2_ensembl_np.select('#Uploaded_variation', 'Consequence')
new
```

```python
def label_consequence(consequence):
    if 'stop_gained' in consequence:
        return 'stop_gained'
    elif 'missense_variant' in consequence:
        return 'missense_variant'
    elif 'frameshift_variant' in consequence:
        return 'frameshift_variant'
    elif 'inframe_deletion' in consequence:
        return 'inframe_deletion'
```

```python
try1 = new.apply(label_consequence, 'Consequence')
new2 = new.with_column('Our Consequences', try1)
new2
yay = new2.where('Our Consequences', are.not_equal_to(None))
test3 = yay.group('Our Consequences')
test3
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.pie(test3.column('count'), autopct='%1.1f%%', shadow = True)
#,labels = test3.column('Our Consequences'), autopct='%1.1f%%', shadow = True)
plt.legend(title = 'MYL2 Consequences:', fontsize="9", labels = test3.column('Our Consequences'), loc ="lower center", bbox_to_anchor=(0.37, -0.2, 0.2, 0.2))
```

```python
MYL2_VUS = pd.read_csv('MYL2_VUS.txt', sep='\t')
newMYL2_VUS = MYL2_VUS['Condition(s)'].str.contains("Hypertrophic cardiomyopathy")
MYL2_VUS1 = MYL2_VUS[newMYL2_VUS].dropna(subset=['Protein change'])
MYL2_VUS1
```

```python
num_amino_acids_myl2 = 166

count = [0] * num_amino_acids_myl2

for i in MYL2_combined.index:
    s = MYL2_combined['Protein change'][i].replace('*', '')
    x = "".join([ele for ele in s if ele.isnumeric()])
    if x:
        x_value = int(x)
        count[x_value] += 1
        
    
# x axis values
x = [i for i in range(num_amino_acids_myl2)]
# corresponding y axis values
y = count
  
# plotting the points 
plt.plot(x, y)
  
# naming the x axis
plt.xlabel('Residue number')
# naming the y axis
plt.ylabel('Frequency')
  
# giving a title to my graph
plt.title('Frequencies of mutations at each residue MYL2')
  
# function to show the plot
plt.show()

# save figure
# plt.savefig('mybpc3.png')

plt.close()
```

```python
f = open("mybpc3_mutationfreq.txt", "w")
f.write("attribute: position\n")
f.write("recipient: residues\n")
for i in range(num_amino_acids_myl2):
    f.write("\t" + ":" + str(i+1) + "\t" + str(count[i]) + "\n")
f.close()
```
