```python
MYBPC3_Path = pd.read_csv('MYBPC3NN.txt', sep='\t')
newMYBPC3_Path = MYBPC3_Path['Condition(s)'].str.contains("Hypertrophic cardiomyopathy")
MYBPC3_Path1 = MYBPC3_Path[newMYBPC3_Path].dropna(subset = ['Protein change'])
MYBPC3_Path1
```

```python
MYBPC3_LP = pd.read_csv('MYBPC3_LP.txt', sep='\t')
newMYBPC3_LP = MYBPC3_LP['Condition(s)'].str.contains("Hypertrophic cardiomyopathy")
MYBPC3_LP1 = MYBPC3_LP[newMYBPC3_LP].dropna(subset=['Protein change'])
MYBPC3_LP1
```

```python
combined1 = [MYBPC3_LP1, MYBPC3_Path1]
MYBPC3_combined = pd.concat(combined1)
MYBPC3_combined

#pd.set_option('display.max_rows', None)
#MYBPC3_combined
```
